## webscraping

```python
import requests ## requisição
import pandas as pd
from bs4 import BeautifulSoup ## manipulação de html
```

```python
url = 'https://www.formula1.com/en/results.html/1985/races.html'
```

```python
## Requisição do site
response = requests.get(url).text
```

```python
## começa a tratar os dados
Soup = BeautifulSoup(response, "html.parser")
```

```python
## Filtrar os dados em uma tabela - apenas a tabela dos resultados
table = Soup.find('table',{'class':'resultsarchive-table'})
```
